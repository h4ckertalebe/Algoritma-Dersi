```python
import random

calisanlar = [
    {"isim": "Mete", "maas": random.randint(3000, 8500)},
    {"isim": "Alptug", "maas": random.randint(3000, 8500)},
    {"isim": "Sevgi", "maas": random.randint(4500, 5000)}
]

zam_orani = 0.10  # %10 zam
print("Çalışan Maaş Analizi ve Zam Raporu:")
for calisan in calisanlar:
    if calisan["maas"] < 8000:
        print(f"Uyarı: {calisan['isim']} adlı çalışanın maaşı 8000 TL'yi aştığı için işi sonlandırılıyor.")
        break

    if calisan["maas"] < 4000:
        print(f"{calisan['isim']} adlı çalışanın maaşı {calisan['maas']} TL'den düşük olduğu için zam yapılmayacaktır.")

    if calisan["maas"] >= 5000:
        calisan["maas"] += calisan["maas"] * zam_orani
        print(f"{calisan['isim']} adlı çalışanın zamlı maaşı: {calisan['maas']} TL (5000 TL'den fazla)")
    else:
        pass  # Geçici olarak işlem yapılmayan durumlar için pass kullanılır

print("Analiz tamamlandı.")
```

```python
# Gerekli kütüphanelerin yüklenmesi
!pip install geoip2 requests pandas matplotlib seaborn

import requests
import json
import pandas as pd
import geoip2.database
import matplotlib.pyplot as plt
import seaborn as sns

# GeoLite2 veritabanı dosya yolu (Bu yolu kendi dosya yolunuzla değiştirin)
geoip_db_path = 'GeoLite2-City.mmdb'

# JSON dosyasını oluştur veya temizle
with open('output.json', 'w') as f:
    f.write('')

# 5 sayfa dolaş ve verileri çek
for page in range(1, 6):  # 1'den 5'e kadar sayfaları dolaş
    # Her sayfadaki veriyi çek
    response = requests.get(
        f'https://www.usom.gov.tr/api/address/index?type=ip&page={page}',
        headers={'accept': 'application/json'}
    )

    # Veriyi JSON dosyasına ekle
    with open('output.json', 'a') as f:
        json.dump(response.json(), f)
        f.write('\n')  # Her sayfa arasına boşluk ekle

print("Veriler output.json dosyasına kaydedildi.")

# output.json dosyasını oku
with open('output.json', 'r') as f:
    # Her bir satırı (sayfa) oku ve JSON olarak parse et
    json_lines = f.readlines()

# Boş bir data sözlüğü oluştur
data = {
    "id": [],
    "url": [],
    "type": [],
    "desc": [],
    "source": [],
    "date": [],
    "criticality_level": [],
    "connectiontype": []
}

# Her bir sayfadaki veriyi işle
for line in json_lines:
    page_data = json.loads(line)  # JSON satırını parse et
    for model in page_data["models"]:  # models listesindeki her bir öğeyi al
        data["id"].append(model["id"])
        data["url"].append(model["url"])
        data["type"].append(model["type"])
        data["desc"].append(model["desc"])
        data["source"].append(model["source"])
        data["date"].append(model["date"])
        data["criticality_level"].append(model["criticality_level"])
        data["connectiontype"].append(model["connectiontype"])

# Veri çerçevesini oluştur
df = pd.DataFrame(data)

# IP adreslerine karşılık gelen şehir ve ülke bilgilerini bul
def get_location(ip):
    try:
        with geoip2.database.Reader(geoip_db_path) as reader:
            response = reader.city(ip)
            city = response.city.name if response.city.name else "Unknown"
            country = response.country.name if response.country.name else "Unknown"
            return city, country
    except Exception as e:
        print(f"Hata: {e}")
        return "Unknown", "Unknown"

# Şehir ve ülke bilgilerini ekle
df[['city', 'country']] = df['url'].apply(lambda x: pd.Series(get_location(x)))

# Ülke ve şehir bilgisini birleştir
df['location'] = df['country'] + ' - ' + df['city']

# Veriyi görüntüle
print(df.head())

# Kritiklik Seviyesine Göre Ülke ve Şehir Dağılımı
plt.figure(figsize=(14, 8))
sns.countplot(data=df, y='location', hue='criticality_level', palette='Set3')
plt.title('Kritiklik Seviyesine Göre Ülke ve Şehir Dağılımı')
plt.xlabel('Sayı')
plt.ylabel('Ülke - Şehir')
plt.legend(title='Kritiklik Seviyesi', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
